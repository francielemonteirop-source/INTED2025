##Estrutura condicional

if - se

elif - senão se

else - senão

```python
nota = float(input("Digite sua nota: "))
if nota >= 7:
  print("Aprovado")
elif nota >= 5:
  print("Recuperação")
else:
  print("Reprovado")
```

4- Crie um algoritmo que receba uma idade do usuário, e depois verifique em qual faixa etária está á idade:

até 12 - Você é uma criança!

de 13 á  18 - Você é um adolescente!

de 19 á 60 - Você é um adulto!

acima de 60 - Você é idoso!

```python
idade = int(input("Digite sua idade: "))

if idade <= 12:
  print("Você é uma criança!")
elif idade <= 18:
  print("Você é um adolescente!")
elif idade <= 60:
  print("Você é um adulto!")
else:
  print("Você é idoso!")
```
