Peça a quantidade de horas trabalhadas e o valor por hora. Se trabalhou até 40 horas, calcule o salário normalmente (horas × valor por hora). Se trabalhou mais de 40 horas, as horas extras devem ter um adicional de 50% (horas extras × valor por hora × 1.5). Crie um algoritmo em Python que realize esse cálculo corretamente.

```python
horas_trabalhadas = float(input("Digite a quantidade de horas trabalhadas: "))
valor_por_hora = float(input("Digite o valor por hora: "))

if horas_trabalhadas <= 40:
  salario = horas_trabalhadas * valor_por_hora
else:
  horas_normais = 40
  horas_extras = horas_trabalhadas - 40
  salario = (horas_normais * valor_por_hora) + (horas_extras * valor_por_hora * 1.5)

print(f"O salário é: R$ {salario:.2f}")
```

```python
import math

def check_multiples_of_5():
  numero = int(input("Digite um número para verificar se é múltiplo de 5: "))
  if numero % 5 == 0:
    print(f"O número {numero} é múltiplo de 5.")
  else:
    print(f"O número {numero} não é múltiplo de 5.")

def check_even_or_odd():
  numero = int(input("Digite um número para verificar se é par ou ímpar: "))
  if numero % 2 == 0:
    print(f"O número {numero} é par.")
  else:
    print(f"O número {numero} é ímpar.")

def check_prime_number():
  numero_primo = int(input("Digite um número para verificar se é primo: "))
  is_prime = True
  if numero_primo <= 1:
    is_prime = False
  elif numero_primo == 2:
    is_prime = True
  else:
    for i in range(2, int(math.sqrt(numero_primo)) + 1):
      if numero_primo % i == 0:
        is_prime = False
        break
  if is_prime:
    print(f"O número {numero_primo} é primo.")
  else:
    print(f"O número {numero_primo} não é primo.")

def convert_temperature():
  temp_value = float(input("Digite o valor da temperatura: "))
  unit = input("Digite a unidade da temperatura (C para Celsius, F para Fahrenheit): ").upper()
  if unit == 'C':
    fahrenheit = (temp_value * 9/5) + 32
    print(f"{temp_value}°C é igual a {fahrenheit:.2f}°F")
  elif unit == 'F':
    celsius = (temp_value - 32) * 5/9
    print(f"{temp_value}°F é igual a {celsius:.2f}°C")
  else:
    print("Unidade inválida. Por favor, digite 'C' ou 'F'.")


choice = -1 # Initialize choice to a non-zero value to enter the loop

while choice != 0:
  print("\nCalculadora v25")
  print("Digite:")
  print("1 - Múltiplos de 5")
  print("2 - Números pares/ímpares")
  print("3 - Números primos")
  print("4 - Conversor de temperatura")
  print("0 - Sair")
  print("-------------------------------------------------------------------------------")

  try:
    choice = int(input("Selecione uma opção: "))

    if choice == 0:
      print("Saindo da calculadora. Até mais!")
    elif choice == 1:
      check_multiples_of_5()
    elif choice == 2:
      check_even_or_odd()
    elif choice == 3:
      check_prime_number()
    elif choice == 4:
      convert_temperature()
    else:
      print("Opção inválida. Por favor, digite uma opção entre 0 e 4.")
  except ValueError:
    print("Entrada inválida. Por favor, digite um número.")
```
